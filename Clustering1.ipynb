## scaling the data

```python
from sklearn.preprocessing import StandardScaler
mm=StandardScaler()
data=mm.fit_transform(data)
```

```python
data=pd.DataFrame(data,columns=['Alcohol', 'Malic_Acid', 'Ash', 'Ash_Alcanity', 'Magnesium',
       'Total_Phenols', 'Flavanoids', 'Nonflavanoid_Phenols',
       'Proanthocyanins', 'Color_Intensity', 'Hue', 'OD280', 'Proline'])
```

```python
data.head()
```

## K-Means clustering

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmean=KMeans(n_clusters=i,init='k-means++',n_init=20)
  kmean.fit(data)
  wcss.append(kmean.inertia_)
  wcss
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow method')
plt.xlabel("no.of cluters")
plt.ylabel("wcss")
plt.show()
```

```python
kmean=KMeans(n_clusters=3,init='k-means++',random_state=42)
kmean.fit(data)
y_pred=kmean.predict(data)
y_pred
```

```python
from sklearn.metrics import silhouette_score
sil_km=silhouette_score(data,y_pred)
sil_km
```

```python
plt.scatter(data['Flavanoids'],data['OD280'])
plt.xlabel=("Flavanoids")
plt.ylabel=("OD280")
plt.show()
```

```python
plt.scatter(data['Alcohol'],data['Total_Phenols'],c=y_pred)
plt.xlabel=("Alcohol")
plt.ylabel=("Total_Phenols")
plt.show()
```

## Agglomerative Hierarchical Clustering(AHC)

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(data,method='ward'))
plt.title('Dendrogram')

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ach=AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
y_ach=ach.fit_predict(data)
y_ach
```

```python
sil_ach=silhouette_score(data,y_ach)
sil_ach
```

```python
# IN both KMeans and AH clustering we got 3 clusters
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
dbs=DBSCAN(eps=3,min_samples=3)
y_dbs=dbs.fit_predict(data)
```
